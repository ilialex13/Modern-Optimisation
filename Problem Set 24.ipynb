```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

# 1.

```python
N = 2**14
k = 2**13
x = np.zeros(N)
while sum(x!=0) < 50:
    idx = np.random.randint(low=0, high=len(x), size=1)
    p = np.random.uniform()
    if p > 0.5:
        x[idx] = 1
    else:
        x[idx] = -1

A = np.random.normal(0, 1, (k, N))
print(A.shape)

epsilon = np.random.normal(0, np.sqrt(10**-4), k)
print(epsilon.shape)

y = A @ x + epsilon
y.shape
```

# 2.

```python
def proj_l1_ball(v, t):
    if t <= 0:
        return "t must be greater than 0"

    if np.linalg.norm(v, 1) <= t:
        return v

    # create u
    u = abs(v)

    # create mu
    mu = np.sort(u)[::-1]
    
    # find rho
    rho = np.sum([mu[j] - (np.sum(mu[:j+1]) - t)/(j+1) > 0 for j in range(0, len(u))])

    # find theta
    theta = (np.sum(mu[:rho]) - t)/rho
    
    #output w
    w = np.sign(v) * np.where(u - theta < 0, 0, u - theta)
    return w
```

```python
def beta_k(k):
    if k == 0:
        return 0
    else:
        return (k-1)/(k+2)

def f(x, A, y):
    return 0.5 * (np.linalg.norm(y - A @ x)**2)

def grad_f(x, A, y):
    return -A.T @ (y - A @ x)

def nesterov(x_start, t, iterations, A, y):
    x_minus = x_start
    x_k = x_start
    x_k1 = x_start
    alpha = 1
    for k in range(iterations):
        # extrapolation
        y_k = x_k + beta_k(k) * (x_k - x_minus)
        
        # ensure we have a valid alpha
        l_x1_yk = f(y_k, A, y) + (grad_f(y_k, A, y)).T @ (x_k1 - y_k)
        while f(x_k1, A, y) > l_x1_yk + (1/(2*alpha) * (np.linalg.norm(x_k1 - y_k)**2)):
            alpha /= 2  

        # gradient projection
        x_k1 = proj_l1_ball(y_k - alpha * grad_f(y_k, A, y), t)

        # check if converged
        if np.linalg.norm(x_k1 - x_k) < (10**-15):
            print(f"converged in {k} iterations")
            return x_k1

        # update values
        x_minus = x_k
        x_k = x_k1
        
    return x_k1
```

# 3. and 4.

```python
# define initial x
x_start = np.random.randint(low=0, high=10, size=N)
x_out = nesterov(x_start, t=50, iterations=501, A=A, y=y)

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (15,10)

time = np.arange(1, N+1)

figs, axes = plt.subplots(2, 1)

axes[0].scatter(time, x, color='blue')
axes[1].scatter(time, x_out, color='orange')
plt.xlabel('time')
plt.ylabel('Signal')
plt.title('Original (Blue) vs Recovered (Orange) Signal, t=50')
```

Using t=30, it is clear that we are unable to recreate the signal completely, with not all information being able to be captured due to our sparsity estimation being smaller than the true sparsity, resulting in a subpar estimation.
